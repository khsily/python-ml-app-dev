### 2. 붓꽃 데이터를 사용해 머신러닝 하기
1. 붓꽃 데이터를 내려받은 'iris.csv'를 읽어 들입니다.
2. 읽어 들인 붓꽃 데이터에서 꽃받침과 꽃잎의 너비와 길이 정보(데이터 부분), 봇꽃 풍종 정보(레이블 부분)를 분리합니다.
3. 전체 데이터 중에서 80%를 학습 전용 데이터 20%를 테스트 전용 데이터로 분리합니다.
4. 학습 전용 데이터를 사용해 학습하고 테스트 전용 데이터를 사용해 품종 분류를 제대로 하는지 확인합니다.

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# 붓꽃 데이터 읽어 들이기
iris_data = pd.read_csv("data/iris.csv", encoding="utf-8")

# 붓꽃 데이터를 레이블과 입력 데이터로 분리하기
y = iris_data.loc[:,"Name"]
x = iris_data.loc[:,["SepalLength", "SepalWidth", "PetalLength", "PetalWidth"]]

# 학습 전용과 테스트 전용 분리하기
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, train_size=0.8, shuffle=True)

# 학습하기
clf = SVC(gamma="scale")  # FutureWarning 을 없애기위해 gamma 값을 'scale' 로 설정
clf.fit(x_train, y_train)

# 평가하기
y_pred = clf.predict(x_test)
print("정답률 = ", accuracy_score(y_test, y_pred))
```
